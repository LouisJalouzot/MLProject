```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from sklearn.preprocessing import MinMaxScaler
```

```python
xTrain = pd.read_csv('xTrainMean.csv')
yTrain = pd.read_csv('y_train_CGZ7jXu.csv')
```

```python
yTrain = yTrain.set_index('sample_id')
xTrain = xTrain.set_index('sample_id')
```

```python
xTrain
```

```python
yTrain
```

```python
scaler = MinMaxScaler()
```

```python
df = xTrain.join(yTrain).dropna()
df
```

```python
scaler.fit(df)
sdf = scaler.transform(df)
train = sdf[:,:-1]
tar = sdf[:,-1:]
train.shape, tar.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Input
```

```python
model = Sequential()
model.add(Input(25))
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(25, activation='relu'))
model.add(Dense(1, activation='relu'))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.fit(train, tar, batch_size=100, epochs=10)
```

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
```
